```python
import os
import MDAnalysis as mda
from MDAnalysis.analysis import distances
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
wd = "/home/siu/md/tau/tau_7upe_S603_12points-2/S603_12"
gro = 'md.gro'  # .gro 파일 경로
xtc = 'md1.xtc'  # .xtc 파일 경로
```

```python
os.chdir(wd)
```

```python
# 단백질과 리간드 시스템 불러오기
u = mda.Universe(gro, xtc)

# 단백질과 리간드 선택
protein = u.select_atoms("protein")
ligand = u.select_atoms("resname LIG")  # 리간드 명을 적절히 변경하세요.
```

```python
# 모든 단백질 잔기에 대한 딕셔너리 생성
residue_info = {i: {"index": i, "resname": res.resname, "resid": res.resid} 
                for i, res in enumerate(protein.residues)}
```

```python
# 각 프레임마다 단백질 잔기와 리간드 간의 거리 계산
n_residues = len(protein.residues)
n_frames = len(u.trajectory)

# 거리 정보를 저장할 배열 생성 (residue × frame)
distance_matrix = np.zeros((n_residues, n_frames))
```

```python
for ts in u.trajectory:
    frame_idx = ts.frame
    
    # 리간드의 무게중심 계산
    ligand_com = ligand.center_of_mass()
    
    # 각 잔기와 리간드 간의 거리 계산
    for i, res in enumerate(protein.residues):
        # 잔기의 무게중심 계산
        res_com = res.atoms.center_of_mass()
        # 거리 계산
        dist = np.linalg.norm(res_com - ligand_com)
        distance_matrix[i, frame_idx] = dist
```

```python
# 10 Å 이하를 녹색(접촉), 그 외는 흰색으로 히트맵을 생성
contact_matrix = np.where(distance_matrix <= 10, 1, 0)

# 각 잔기별로 접촉 비율 계산
contact_ratio = np.sum(contact_matrix, axis=1) / n_frames

# 70% 이상 접촉한 잔기는 붉은색으로 표시
colored_matrix = np.copy(contact_matrix)
colored_matrix[contact_ratio >= 0.7, :] = 2  # 2는 붉은색

# 사용자 정의 컬러맵: 흰색(접촉 없음), 녹색(10 Å 이하), 붉은색(70% 이상 접촉)
cmap = ListedColormap(['white', 'green', 'red'])

# interaction이 없는 잔기를 제외
non_zero_indices = np.where(np.sum(contact_matrix, axis=1) > 0)[0]
filtered_matrix = colored_matrix[non_zero_indices, :]

# 잔기 이름과 번호, 인덱스 번호 및 접촉 비율로 y축 레이블 생성
y_labels = [
    f"{protein.residues[i].resname}{protein.residues[i].resid} ({i}) / {contact_ratio[i]*100:.1f}%" 
    for i in non_zero_indices
]

# 두 번째 히트맵 그리기 (interaction이 없는 잔기 제외, y축 레이블 포함)
plt.figure(figsize=(10, 8))
plt.imshow(filtered_matrix, aspect='auto', cmap=cmap, interpolation='nearest')
plt.colorbar(label='Interaction: Green (≤10Å), Red (≥70% frames)')
plt.xlabel('Frame')
plt.ylabel('Residue (Amino Acid + Number + Index) / Ratio (%)')
plt.yticks(ticks=np.arange(len(non_zero_indices)), labels=y_labels)
plt.title('Protein-Ligand Distance Interaction Heatmap (Filtered)')
plt.show()
```

```python
# 접촉 비율별로 내림차순으로 정렬하기 위한 데이터 준비
sorted_indices = np.argsort(contact_ratio[non_zero_indices])[::-1]
sorted_labels = [y_labels[i] for i in sorted_indices]
sorted_ratios = contact_ratio[non_zero_indices][sorted_indices] * 100  # 비율을 퍼센트로 변환

# 바 그래프 그리기
plt.figure(figsize=(10, 8))
plt.barh(sorted_labels, sorted_ratios, color='skyblue')
plt.xlabel('Interaction Ratio (%)')
plt.ylabel('Residue (Amino Acid + Number + Index)')
plt.title('Protein-Ligand Interaction Ratio (Sorted by %)')
plt.gca().invert_yaxis()  # y축을 내림차순으로
plt.show()
```
