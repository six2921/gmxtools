```python
from rdkit import Chem
```

```python
smarts_pattern = "c1*ccc1"
pattern = Chem.MolFromSmarts(smarts_pattern)
pattern
```

```python
cpd_list = ["C1=CNCC=C1", "C1=COC(=C1)C=O", "C1=COC(=C1)CNC2=NC=NC3=C2NC=N3", "C1=C(OC(=C1)C=O)CO", 
             "C1=CC=C2C(=C1)C=CN2", "C1=CNC=C1", "C1=CC=C2C(=C1)C3=CC=CC=C3N2", "CCN1C(=O)C=CC1=O", 
             "CN1C=CC=C1C=O", "C1=CN(C=C1)CC2=CC=CO2", "CC(=O)OC1=CC=CC=C1C(=O)O", "CCCC(C)(COC(=O)N)COC(=O)N"]
```

```python
# cpd_list 보여주기
molecules = [Chem.MolFromSmiles(smiles) for smiles in cpd_list]
img = Draw.MolsToGridImage(molecules, molsPerRow=10, subImgSize=(100, 100), useSVG=True)
img
```

```python
# SMILES에 대해 SMARTS 패턴 매칭 결과를 저장할 딕셔너리
results = {}

# 각 SMILES에 대해 패턴 매칭 테스트
for smiles in cpd_list:
    mol = Chem.MolFromSmiles(smiles)
    if mol and mol.HasSubstructMatch(pattern):
        results[smiles] = True
    else:
        results[smiles] = False
```

```python
# SMILES 리스트를 RDKit 분자 객체로 변환
true_smiles = [smiles for smiles, match in results.items() if match][:10]
true_molecules = [Chem.MolFromSmiles(smiles) for smiles in true_smiles]

# True인 SMILES의 그리드 이미지 생성 및 저장
true_img = Draw.MolsToGridImage(true_molecules, molsPerRow=10, subImgSize=(100, 100), useSVG=True)
print(true_smiles)
true_img
```

```python
# SMILES 리스트를 RDKit 분자 객체로 변환
false_molecules = [Chem.MolFromSmiles(smiles) for smiles in false_smiles]
false_smiles = [smiles for smiles, match in results.items() if not match][:10]

# False인 SMILES의 그리드 이미지 생성 및 저장
false_img = Draw.MolsToGridImage(false_molecules, molsPerRow=10, subImgSize=(100, 100), useSVG=True)
print(false_smiles)
false_img
```
